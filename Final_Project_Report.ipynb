**4.1 Answer:**  
Our overall framework followed a structured machine learning process, beginning with framing the business problem and performing exploratory data analysis (EDA). After extracting foundational information from the dataset, we cleaned and preprocessed the data, conducted feature engineering, and trained and fine-tuned multiple models. The entire process was iterative, so decisions made at each stage informed the next. To monitor generalization and prevent overfitting, we used RepeatedKFold cross-validation (5 folds, 5 repeats) consistently across baseline evaluation, feature engineering, feature selection, and hyperparameter tuning. We selected Mean Absolute Error (MAE) as the primary error metric because it is directly interpretable as average prediction error in dollars and is less sensitive to extreme outliers such as high-value properties than RMSE would be. The mean and standard deviation of CV MAE were reported at each stage to assess both accuracy and stability. RMSE and R² were additionally reported in the final evaluation.


### 4.2 Data Cleaning and Preprocessing

**4.2 Answer:**  
The raw data contained a large number of missing values, redundant features, and heavy right-skew in the target variable (`taxvaluedollarcnt`) partly driven by extreme high-value property outliers. Several columns were dropped as unsuitable for regression regardless of having missing values: `parcelid` is a unique row identifier with no predictive value, `assessmentyear` is nearly constant across all rows,  `rawcensustractandblock` and `censustractandblock` are geographic identifiers unsuitable as continuous predictors, and `calculatedbathnbr`. Features missing more than 70% of their values were also dropped, including `storytypeid`, `architecturalstyletypeid`, and `buildingclasstypeid`. The 70% threshold was chosen to remove near-empty columns while retaining potentially useful features. Row samples were dropped due to three reasons : rows where the target was null as they cannot be used for regression training, rows missing more than 50% of remaining features, and rows with extreme outliers in the target identified using the 3×IQR rule, which could likely distort model training. For remaining missing values, three imputation strategies were applied by feature type: binary features were filled with zero, numerical features with the column median, and categorical features with the column mode. The remaining categorical features after cleaning, such as propertyzoningdesc, had too many unique values to encode practically via one-hot encoding. No suitable alternative encoding strategy was identified, so they were dropped before modeling.

### 4.4 Model Selection

Describe how you selected your model.

Include:
- models you tried and why
- how you evaluated generalization
- hyperparameter tuning
- how you chose the final model

**4.4 Answer:**  
Three models were selected for this project: Linear Regression, Decision Trees, and Random Forest. Linear Regression was used as a simple baseline, Decision Tree was used to test a non-linear model capable of complex splits, and Random Forest as an ensemble method expected to outperform both through variance reduction. All models were evaluated using repeated 5-fold cross-validation with 5 repeats at every stage, with a held-out test set reserved exclusively for final evaluation to prevent data leakage.

Random Forest achieved the lowest mean CV MAE (151,411), followed by Linear Regression (171,387), and Decision Tree last (196,707) due to severe overfitting at default settings. Feature engineering did not significantly improve any model. Feature selection helped most for the Decision Tree, dropping its MAE to 163,958 by reducing unnecessary splits, while Random Forest was largely unaffected since it performs internal feature selection through random feature subsampling. Linear Regression was replaced by Ridge for the tuning stage, as Linear Regression alone has no meaningful hyperparameters to tune, and Ridge's alpha parameter provided a practical tuning opportunity while remaining within the linear model family. For hyperparameter tuning, GridSearchCV was used for Ridge Regression (tuning alpha from 0.01 to 1000, best value 100) and Decision Tree (tuning max_depth, min_samples_split, and min_samples_leaf). RandomizedSearchCV was used for Random Forest given its larger hyperparameter space, tuning n_estimators, max_depth, and max_features. Tuning yielded modest but consistent improvements across all three models. The final model selected was the tuned Random Forest with n_estimators=100, max_depth=20, and max_features=None, achieving a test set MAE of 147,791 and R-squared of 0.5186. This decision was based on its consistently superior performance across all stages of the pipeline. The key trade-off considered was interpretability versus performance. Although Ridge Regression is more interpretable, Random Forest's substantially lower error and ability to capture non-linear relationships between property size, location, and value made it the ideal choice.

**5.3 Answer:**  

The residuals from the final random forest model reveal consistent patterns that highlight where the model performs well and where it struggles.

### Residual distribution and central bias
The residual histogram is sharply centered at zero and roughly symmetric, indicating that on average the model is unbiased. However, the distribution has heavy tails on both sides, meaning a meaningful subset of properties have large errors in either direction. This explains why RMSE ($205,261) is significantly larger than MAE ($147,791) — a relatively small number of large errors disproportionately inflate squared-error metrics.

**Systematic underprediction of high-value properties.**  
The Predicted vs. Actual plot shows that for properties above roughly $1M, predictions consistently fall **below** the diagonal. The model underpredicts expensive homes because:
- High-value properties are underrepresented in the training data, so tree splits are driven by the more common mid-range homes.
- Random Forest predictions are bounded by the average of leaf values from training data, making it inherently conservative at the upper extreme of the target distribution.

**Systematic overprediction of low-value properties.**  
Similarly, there is an observable trend of properties with an actual value in the $100,000 - $250,000 range that are overpredicted by the model, sometimes upwards of $500,000. As discussed in 5.2, this finding may reflect limitations of the model in capturing location and condition factors that cap the property value at a low level.

**Directional bias across the price range.**  
The Residuals vs. Predicted plot shows that residual *spread* remains roughly stable across predicted values, but the *sign* of the error shifts systematically: positive residuals dominate at the low end (the model underpredicts homes that are actually mid-priced) and negative residuals dominate at the high end (the model overpredicts homes labeled as high-value). This indicates Random Forest is regressing predictions toward the training mean at both extremes — a known limitation when the target distribution is skewed. Applying a **log transformation** to the target could help reduce this directional bias.

**Outliers and unusual observations.**  
A small number of properties show extreme residuals — both positive (severe underpredictions on luxury homes) and negative (overpredictions on properties whose recorded tax value seems unusually low for their features). Some of these likely reflect data quality issues (e.g., assessed values that lag market value, special-use properties, or data entry inconsistencies) rather than genuine model failure.

**What's surprising and worth improving.**  
The most striking observation is the structural bias at both price extremes despite Random Forest's overall strong performance and well-centered residuals. Possible improvements include:
- Applying a **log transformation** to `taxvaluedollarcnt` to compress the upper tail of the target distribution.
- Training **separate models** for distinct price tiers (e.g., entry-level, mid-market, luxury).
- Adding richer **location-based features** (neighborhood-level medians, school ratings, proximity to amenities) to better differentiate high-value areas.
- Trying **gradient-boosted trees** (XGBoost, LightGBM), which often handle skewed targets and tail behavior more effectively than Random Forest.

Overall, the model is reliable and well-calibrated for typical homes but loses accuracy at both extremes of the market — an important caveat for any business application focused on luxury or distressed property pricing.

**6.1 Answer:**  

### Main Findings

This project set out to build a reliable model for predicting residential property values from the Zillow housing dataset, identify the features that most influence those values, and produce a pipeline that generalizes well to new properties. After comparing Linear/Ridge Regression, Decision Trees, and Random Forest across baseline, feature engineering, feature selection, and hyperparameter tuning stages, the tuned **Random Forest** emerged as the strongest model with a test MAE of **$147,791**, RMSE of **$205,261**, and R² of **0.519**. The close agreement between cross-validated MAE ($151,156) and test MAE confirms that the model generalizes well rather than overfitting. Feature importance analysis consistently showed that **square footage, geographic features (latitude, longitude, regionidzip), and year built** are the dominant drivers of assessed property value, directly addressing our original objective of identifying the most important predictors.

### Business and Practical Implications

For a real estate company, these results offer several actionable takeaways. First, accurate, automated property valuation is achievable with a relatively small set of high-quality predictors, reducing reliance on manual comparisons. Second, the model is most reliable for typical mid-market homes (roughly $200K–$800K), making it well suited for bulk valuation tasks, market analysis, and initial pricing recommendations. Third, the strong influence of geographic features confirms that **location data quality is a high-leverage investment** — improvements in neighborhood-level features will likely produce larger gains than further model tuning.

### Limitations and Constraints

Several limitations should be acknowledged. The model systematically **underpredicts high-value properties** (above ~\$1M) and **overpredicts very low-value properties**, indicating it struggles at both tails of the price distribution. The R² of 0.519 means roughly half of the variance in property values remains unexplained, likely due to missing information about property condition, neighborhood desirability, school quality, and recent market trends. Additionally, several high-cardinality categorical features (e.g., `propertyzoningdesc`) had to be dropped because no practical encoding strategy was available, which may have removed useful signal. The dataset is also a static snapshot, so the model cannot capture temporal market dynamics.

### Future Directions

Future work could improve performance and robustness by:
- Applying a **log transformation** to the target to better handle the heavy right tail of property values
- Training **tier-specific models** (entry-level, mid-market, luxury) to address bias at the price extremes
- Incorporating **external data sources** such as school ratings, crime statistics, walkability scores, and neighborhood-level economic indicators
- Testing **gradient-boosted models** (XGBoost, LightGBM), which often handle skewed targets more effectively than Random Forest
- Exploring **target encoding** for high-cardinality categorical features that were dropped

### Final Recommendation

We recommend deploying the tuned Random Forest model as a **first-pass valuation tool** for mid-market residential properties, where its accuracy is strongest and most stable. For high-value or unusually low-value properties, predictions should be flagged for **human expert review** rather than used directly. In parallel, the company should invest in enriching the dataset with location-based and condition-related features, which our analysis identifies as the most promising path to meaningful accuracy gains. This combination of automated valuation for typical cases and expert oversight for edge cases provides an immediate, practical improvement over traditional valuation methods while leaving a clear roadmap for future enhancement.
